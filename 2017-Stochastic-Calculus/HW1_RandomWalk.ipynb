```python
import random
import matplotlib.pyplot as plt
from math import sqrt

print ('Write N : ')
N = input()
N = int(N)
T = 1
X = [0]

def random_walk(T, N):
    y = 0

    case = {1: y + sqrt(T / N),
            2: y - sqrt(T / N)
            }

    for _ in range(N):
        y += case[random.randint(1, 2)]
        X.append(y)


def plot_graph(T, X):
    plt.plot(T, X)
    plt.xlabel('T')
    plt.ylabel('X')
    plt.show()
    
Time = [i*T/N for i in range(N+1)]

random_walk(T, N)
plot_graph(Time, X)
```
